```python
def create_date_list(start_year=2020, end_year=2022):
    dates = []
    while start_year < end_year+1:
        start_date = str(start_year)+'-05-01'
        end_date = str(start_year)+'-08-30'
        dates.append((start_date, end_date))
        start_year += 1
    return dates
```

```python
def maskSrClouds(image):
    # Bit 0 - Fill
    # Bit 1 - Dilated Cloud
    # Bit 2 - Cirrus
    # Bit 3 - Cloud
    # Bit 4 - Cloud Shadow
    qaMask = image.select('QA_PIXEL').bitwiseAnd(int('11111', 2)).eq(0)
    saturationMask = image.select('QA_RADSAT').eq(0)

    return image.updateMask(qaMask).updateMask(saturationMask)
```

```python
summer_lsts = ee.Image()
lst_bands = []
for dates in create_date_list(2020, 2022):
    lst_bands.append(('lst_'+dates[0])[:8])
    summer_lst = ee.ImageCollection("LANDSAT/LC08/C02/T1_L2")\
    .filterDate(ee.Date(dates[0]), ee.Date(dates[1]))\
    .filterBounds(roi)\
    .filter(ee.Filter.lt('CLOUD_COVER', 50))\
    .map(maskSrClouds)\
    .select('ST_B10')\
    .median()\
    .multiply(0.00341802).add(149.0).subtract(273.15)\
    .clip(roi)\
    .rename(('lst_'+dates[0])[:8])
    summer_lsts = summer_lsts.addBands(summer_lst)
summer_lsts = summer_lsts.select(lst_bands)
```

```python
landsat_images = ee.Image()

for dates in create_date_list(2020, 2022):
    landsat_image = ee.ImageCollection("LANDSAT/LC08/C02/T1_L2")\
    .filterDate(ee.Date(dates[0]), ee.Date(dates[1]))\
    .filterBounds(roi)\
    .filter(ee.Filter.lt('CLOUD_COVER', 50))\
    .map(maskSrClouds)\
    .median()\
    .select('SR_B.')\
    .multiply(0.0000275).add(-0.2)\
    .clip(roi)
    landsat_image = landsat_image.rename(['{}_{}'.format(bandName, dates[0][:4]) for bandName in landsat_image.bandNames().getInfo()])


    landsat_images = landsat_images.addBands(landsat_image)

landsat_images = landsat_images.select('SR.*')
```

<h3 style="display:inline;">The mean LST value in the urban area is definitely higher than the rural areas, </h3><h2 style="display:inline; font-weight:bold;">but </h2><h3 style="display:inline;">is the difference in mean LST values in the urban and rural area statistically significant?</h3>
<h4>t-test and Mann-Whitney U test to compare the mean LST values in the rural and urban area</h4>
<li>Null Hypothesis (HO): The mean LST value in urban area is not significantly different from the mean LST value in the rural area</li>
<li>Null Hypothesis (HA): The mean LST value in urban area is significantly different from the mean LST value in the rural area</li>
<li>Significance level (α): Willing to accept a 5% chance of rejecting the HO when it is actually true, therefore, α = 0.05</li>

```python
from scipy.stats import ttest_ind, shapiro, bartlett, mannwhitneyu
```

```python
urban_mean_lst_samples= mean_lst_image.sample(urban_roi, scale = 30, numPixels= 300, seed=42)
rural_mean_lst_samples= mean_lst_image.sample(rural_roi, scale = 30, numPixels= 300, seed=42)

urban_mean_lst_samples = geemap.ee_to_pandas(urban_mean_lst_samples)
rural_mean_lst_samples = geemap.ee_to_pandas(rural_mean_lst_samples)
```

```python
mean_lst_df = pd.concat([urban_mean_lst_samples.add_prefix('urban_'), rural_mean_lst_samples.add_prefix('rural_')], axis=1)
```

```python
mean_lst_df.describe()
```

```python
stat1, p1 = shapiro(urban_mean_lst_samples)
stat2, p2 = shapiro(rural_mean_lst_samples)

print("Sample 1 normality test: statistic = {}, p-value = {}".format(stat1, p1))
print("Sample 2 normality test: statistic = {}, p-value = {}".format(stat2, p2))
```

```python
stat3, p3 = bartlett(urban_mean_lst_samples['lst'], rural_mean_lst_samples['lst'])

print("Bartlett's test: statistic = {}, p-value = {}".format(stat3, p3))
```

```python
t_stat, p_val = ttest_ind(urban_mean_lst_samples, rural_mean_lst_samples, equal_var=False)

print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")

if p_val < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

```python
u, p4 = mannwhitneyu(urban_mean_lst_samples, rural_mean_lst_samples)

print("Mann-Whitney U test results: U = {:.2f}, p = {:.2e}".format(u[0], p4[0]))

if p4[0] < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

<h2>Results</h2>
<h3>Conclusion: Considering the t-test and Mann-Whitney U test results, we can reject the null hypothesis, meaning, the mean LST values in urban area are higher and the difference between mean LST values in the rural area and urban is statistically significant </h3>
<p>
The t-statistic indicates how far the sample means of urban and rural areas are from each other, relative to the spread or variability of the data within each area. Specifically, it indicates the number of standard deviations that the difference between the sample means is from zero, assuming that the null hypothesis (i.e., the mean LST value in urban area is not significantly different from the mean LST value in the rural area) is true. In this case, the t-statistic is 24.076, which is a large number, indicating a significant difference between the two samples.</br>

The p-value is the probability of obtaining a t-statistic as large as the one we observed here, or even larger, assuming that the null hypothesis is true. A p-value of 0.05 or smaller is typically considered statistically significant, which means that there is strong evidence against the null hypothesis. In this case, the p-value is very small (2.94e-88), which means that it is extremely unlikely that we would observe such a large difference between the sample means if the null hypothesis were true. Therefore, we can reject the null hypothesis and conclude that there is a significant difference in mean LST values between the urban and rural areas in the study.</br>

The Mann-Whitney U test outputted a U statistic of 83265.00. This statistic indicates the rank-sum of the two sample datasets being compared. It represents the sum of the ranks of one group (either urban or rural) and can range from 0 to the maximum possible sum of the ranks of both groups.</br>

In this case, since the U value is very high and the p-value is very low, it indicates that there is a significant difference between the mean LST values of the urban and rural areas. Specifically, the U statistic indicates that the rank-sum of the rural group was higher than the rank-sum of the urban group. This suggests that, on average, the rural areas had lower LST values compared to the urban areas.
